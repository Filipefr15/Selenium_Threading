```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import threading
import time
from shutil import which

# A classe Service é usada para iniciar uma instância do Chrome WebDriver
service1 = Service(executable_path=which('chromedriver'))
service2 = Service(executable_path=which('chromedriver'))
service3 = Service(executable_path=which('chromedriver'))
service4 = Service(executable_path=which('chromedriver'))

# webdriver.ChromeOptions é usado para definir a preferência para o browser do Chrome
options = webdriver.ChromeOptions()
#faz com que nao abra a página.
options.add_argument("--headless=new")
options.add_experimental_option('excludeSwitches', ['enable-logging'])

# basicamente o driver É o google chrome.
driver1 = webdriver.Chrome(service=service1, options=options)
driver2 = webdriver.Chrome(service=service2, options=options)
driver3 = webdriver.Chrome(service=service3, options=options)
driver4 = webdriver.Chrome(service=service4, options=options)

#o link que será acessado (uau)
url = 'https://cvmweb.cvm.gov.br/SWB/default.asp?sg_sistema=fundosreg'

all_names = []

def get_url(driver, url, cnpj):
    driver.get(url)
    iframe = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.TAG_NAME, "frame")))
    driver.switch_to.frame(iframe[1])

    inserir_cnpj = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="txtCNPJNome"]')))
    inserir_cnpj.clear()
    inserir_cnpj.send_keys(cnpj)

    button_continuar = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="btnContinuar"]')))
    button_continuar.click()

    driver.switch_to.default_content()
    frame = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@name="Main"]')))
    driver.switch_to.frame(frame)

    checking = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="ddlFundos__ctl0_lnkbtn1"]')))
    #print(f"CNPJ {cnpj}: {checking.text}")
    all_names.append(checking.text)
    driver.quit()

print("Threads are starting")

thread1 = threading.Thread(target=get_url, args=(driver1, url, "13.077.415/0001-05"))
thread2 = threading.Thread(target=get_url, args=(driver2, url, "13.077.418/0001-49"))
thread3 = threading.Thread(target=get_url, args=(driver3, url, "00.852.311/0001-89"))
thread4 = threading.Thread(target=get_url, args=(driver4, url, "23.434.820/0001-97"))


# Start the threads
thread1.start()
thread2.start()
thread3.start()
thread4.start()


# The main thread waits for both threads to finish
thread1.join()
thread2.join()
thread3.join()
thread4.join()

print(all_names)

print("Both threads have finished.")
```
